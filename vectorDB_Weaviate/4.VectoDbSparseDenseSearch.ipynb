```python
# Download the data
resp = requests.get('https://raw.githubusercontent.com/weaviate-tutorials/quickstart/main/data/jeopardy_tiny.json')
data = json.loads(resp.text)  # Load data

# Parse the JSON and preview it
print(type(data), len(data))

def json_print(data):
    print(json.dumps(data, indent=2))

json_print(data[0])
```

```python
# resetting the schema. CAUTION: This will delete your collection
if client.schema.exists("Question"):
    client.schema.delete_class("Question")

class_obj = {
    "class": "Question",
    "description": "A collection of Jeopardy questions",
    "properties": [
        {
            "name": "Category",
            "dataType": ["text"],
            "description": "The category of the question"
        },
        {
            "name": "Question",
            "dataType": ["text"],
            "description": "The text of the question"
        },
        {
            "name": "Answer",
            "dataType": ["text"],
            "description": "The answer to the question"
        }
    ]
}

client.schema.create_class(class_obj)
```

```python
questions = [item['Question'] for item in data]
embeddings = model.encode(questions)
```

```python
# print sample sentence and its embedding
print(questions[0])
print(embeddings[0])
```

```python
# Configure batch
client.batch.configure(batch_size=10)

# Batch import all objects
with client.batch as batch:
    for i, item in enumerate(data):
        properties = {
            "Category": item["Category"],
            "Question": item["Question"],
            "Answer": item["Answer"]
        }
        # Perform data insert with embedding
        client.batch.add_data_object(
            class_name="Question",
            data_object=properties,
            vector=embeddings[i]  # Use the precomputed embedding
        )
```

```python
count = client.query.aggregate("Question").with_meta_count().do()
json_print(count)
```

```python
# write a query to extract the vector for a question
result = (client.query
          .get("Question", ["category", "question", "answer"])
          .with_additional("vector")
          .with_limit(1)
          .do())

json_print(result)
```

## Sparse & Dense Search

```python
# DENSE SEARCH
# Generate embedding for the search concept
search_concept = "animal"
search_vector = model.encode([search_concept])[0]

# Perform Dense Search
response = (
    client.query
    .get("Question", ["question", "answer"])
    .with_near_vector({"vector": search_vector.tolist()})
    .with_limit(3)
    .do()
)

json_print(response)
```

```python
# SPARSE SEARCH
# Define a helper function to print JSON
def json_print(data):
    print(json.dumps(data, indent=2))

# Perform Sparse Search using BM25
response = (
    client.query
    .get("Question", ["question", "answer"])
    .with_bm25(query="animal")
    .with_limit(1)
    .do()
)

json_print(response)
```

```python
# HYBRID SEARCH
# Generate embedding for the search concept
search_concept = "animal"
search_vector = model.encode([search_concept])[0]

# Perform Hybrid Search
response = (
    client.query
    .get("Question", ["question", "answer"])
    .with_hybrid(query="animal", alpha=0.5, vector=search_vector.tolist())
    .with_limit(3)
    .do()
)

json_print(response)
```

```python
# Perform Hybrid Search
response = (
    client.query
    .get("Question", ["question", "answer"])
    .with_hybrid(query="animal", alpha=0, vector=search_vector.tolist())
    .with_limit(3)
    .do()
)

json_print(response)
```

```python
# Perform Hybrid Search
response = (
    client.query
    .get("Question", ["question", "answer"])
    .with_hybrid(query="animal", alpha=1, vector=search_vector.tolist())
    .with_limit(3)
    .do()
)

json_print(response)
```
